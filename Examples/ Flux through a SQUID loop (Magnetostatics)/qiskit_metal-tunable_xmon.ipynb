```python
import qiskit_metal as metal
from qiskit_metal import designs, draw
from qiskit_metal import MetalGUI, Dict, open_docs
from qiskit_metal.qlibrary.qubits.transmon_cross_fl import TransmonCrossFL
from qiskit_metal.qlibrary.terminations.launchpad_wb import LaunchpadWirebond
from qiskit_metal.qlibrary.tlines.straight_path import RouteStraight
```

```python

# Each time you create a new quantum circuit design,
# you start by instantiating a QDesign class.

# The design class `DesignPlanar` is best for 2D circuit designs.
```

```python
design = designs.DesignPlanar()
gui = MetalGUI(design)

# cpw width and dielectric gap size
t_width = '6um'
t_gap="4um"

xmon_options = dict(fl_options = dict(t_width=t_width,t_gap=t_gap)
                   )
qubit = TransmonCrossFL(design, 'qubit', options=xmon_options)


launch_options = dict(pos_x='-600um', pos_y='-600um', orientation='0',trace_width=t_width,trace_gap=t_gap)

lp = LaunchpadWirebond(design, 'launch_pad', options = launch_options)

route_opts = Dict(
    pin_inputs=Dict(
        start_pin=Dict(
            component='launch_pad',
            pin='tie'          # LaunchpadWirebond output pin
        ),
        end_pin=Dict(
            component='qubit',
            pin='flux_line'    # Flux pin on TransmonCrossFL
        )
    ),

    # CPW geometry (match your launchpad + flux stub)
    trace_width=t_width,
    trace_gap=t_gap,
)


flux_route = RouteStraight(
    design,
    'flux_route',
    options=route_opts
)


gui.rebuild()
gui.autoscale()

# gui.rebuild()  # rebuild the design and plot
# gui.autoscale() #resize GUI to see QComponent
```
