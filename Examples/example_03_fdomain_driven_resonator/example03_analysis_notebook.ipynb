## Analyze $S_{21}$ data from a driven resonator simulation to extract resonator parameters

First we need to create a Config object from our already created & simulated AWS Palace configuration file. 

````python
driven_res_config = Config.load_config("config/example03.json")
driven_res_sim = Simulation(driven_res_config,None)                                      
````

Extract $S_{21}$ dataframe from simulation results.

````python
S21_df = driven_res_sim.get_Sij(2,1)
S21_df.head(5)
````

## Extract $f_r$ & $\kappa$ with a Diameter Correction Method (DCM) fit to the data

We can also plot the fit to $|S_{21}|$ data to ensure it looks good

````python
from pypalace.analysis import resonator_analysis 
````

````python
resonator_params = resonator_analysis.get_resonator_parameters_driven(S21_df,show=True,save="Figures/resonator_lineshape.png")
````
