Construct qubit layouts with Quantum (Qiskit) Metal and mesh with pyPalace (Gmsh backend)

```python
import qiskit_metal as metal
from qiskit_metal import designs, draw
from qiskit_metal import MetalGUI, Dict, open_docs
from qiskit_metal.qlibrary.qubits.transmon_cross import TransmonCross
from qiskit_metal.toolbox_python.attr_dict import Dict
from qiskit_metal.qlibrary.core import QComponent
from qiskit_metal.qlibrary.qubits.transmon_pocket import TransmonPocket
```

```python
from pypalace import mesh
```

## Pocket (double pad) Transmon

```python
design = designs.DesignPlanar()
gui = MetalGUI(design)

gui = MetalGUI(design)

# Double-pad qubit (TransmonPocket)
qubit = TransmonPocket(design,
                       "pocket_transmon",
                        
                        # qubit options
                       options=Dict( 
                            pos_x="0mm",
                            pos_y="0mm",
                            orientation="0",
                            pad_width="500um",
                            pad_height="125um",
                            pocket_width="1200um",
                            pocket_height="1200um",
                            pad_gap="30um",
                            # we don't need the quasi-junction for electrostatic / capacitance simulations.
                            inductor_width="0um",
                            
                        # coupler options
                        connection_pads=Dict(
                            readout=Dict(
                                pad_width="143um",
                                pad_height="47um",
                                pad_gap="25um",
                                loc_W = -1,
                                loc_H = +1))))

cpw_width = 0
cpw_gap = 0
cpw_extend = 0
pad_cpw_extent = 0


gui.rebuild()
gui.autoscale()
```

```python
design.qgeometry.tables["poly"]
```

```python
''' Mesh our Quantum Metal design '''

## define which components we want to give mesh attributes ##
Attributes = {"pad_top":1,"pad_bot":2,"readout_connector_pad":3}

mesh_dir = "mesh/"
meshfile = "pocket_transmon.msh"
path_to_mesh = mesh_dir + meshfile

## mesh the design --> returns pandas.dataframe of labeled mesh attributes ##
## look at readthedocs, mesh_Quantum_Metal_design has a lot of flexibility to control surface mesh size, volume mesh size, substrate size, etc ## 
attributes_table = mesh.mesh_Quantum_Metal_design(design,output_mesh=path_to_mesh,Attributes=Attributes,surface_mesh_size=0.0005);
```

## Xmon (Transmon Cross)

```python
design = designs.DesignPlanar()
gui = MetalGUI(design)

## define qubit ##
qubit_options = dict(connection_pads=dict(claw = dict(connector_location = '90', connector_type = '0',claw_length="150um",claw_gap="4um")))
qubit = TransmonCross(design,'qubit',options=qubit_options)

gui.rebuild()
gui.autoscale()
```

```python
design.qgeometry.tables["poly"]
```

```python
''' Mesh our Quantum Metal design '''

## define which components we want to give mesh attributes ##
Attributes = {"cross":1,"claw_connector_arm":2}

mesh_dir = "mesh/"
meshfile = "xmon.msh"
path_to_mesh = mesh_dir + meshfile

## mesh the design --> returns pandas.dataframe of labeled mesh attributes ##
## look at readthedocs, mesh_Quantum_Metal_design has a lot of flexibility to control surface mesh size, volume mesh size, substrate size, etc ## 
attributes_table = mesh.mesh_Quantum_Metal_design(design,output_mesh=path_to_mesh,Attributes=Attributes,surface_mesh_size=0.0005);
```
