```python
import qiskit_metal as metal
from qiskit_metal import designs, draw
from qiskit_metal import MetalGUI, Dict, open_docs
from qiskit_metal.qlibrary.qubits.transmon_cross import TransmonCross
from qiskit_metal.toolbox_python.attr_dict import Dict
from qiskit_metal.qlibrary.core import QComponent
from qiskit_metal.qlibrary.qubits.transmon_pocket import TransmonPocket
```

# Pocket Transmon

```python
design = designs.DesignPlanar()
gui = MetalGUI(design)

gui = MetalGUI(design)

# Double-pad qubit (TransmonPocket)
qubit = TransmonPocket(design,
                       "pocket_transmon",
                        
                        # qubit options
                       options=Dict( 
                            pos_x="0mm",
                            pos_y="0mm",
                            orientation="0",
                            pad_width="500um",
                            pad_height="125um",
                            pocket_width="1200um",
                            pocket_height="1200um",
                            pad_gap="30um",
                            # we don't need the quasi-junction for electrostatic / capacitance simulations.
                            inductor_width="0um",
                            
                        # coupler options
                        # I remove the CPW bit in a CAD software, we don't need the cpw for the capacitance matrix - only qubit + coupler.
                        connection_pads=Dict(
                            readout=Dict(
                                pad_width="143um",
                                pad_height="47um",
                                pad_gap="25um",
                                loc_W = -1,
                                loc_H = +1))))

cpw_width = 0
cpw_gap = 0
cpw_extend = 0
pad_cpw_extent = 0


gui.rebuild()
gui.autoscale()

## save as GDS ##
a_gds = design.renderers.gds
a_gds.export_to_gds("pocket_transmon.gds")
```
